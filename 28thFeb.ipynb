```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import statistics 
from scipy.stats import norm
%matplotlib inline
```

```python
df = pd.read_excel('/content/nbahw.xlsx')
```

```python
df
```

```python
mean = statistics.mean(df['Height'])
```

```python
sd = statistics.stdev(df['Height'])
```

```python
df['Height'].describe()
```

```python
x = np.arange(69, 87, 1)
```

```python
plt.plot(x, norm.pdf(x, mean, sd))
plt.ylabel('Density')
plt.xlabel('Heights')
plt.title('Normal Distribution for Heights')
```

```python
df['Weight'].describe()
```

```python
plt.plot(x, norm.pdf(x, statistics.mean(df['Weight']), statistics.stdev(df['Weight'])))
plt.ylabel('Density')
plt.xlabel('Weights')
plt.title('Normal Distribution for Weights')
```
