6. What is the percentage for Kohli being Not Out.Make a Pie Chart displaying the distribution of Dismissals.



````python
df['Dismissal'].value_counts()
````

````python
caught = df.loc[df['Dismissal'] == 'Caught'].count()[0]
no = df.loc[df['Dismissal'] == 'NO'].count()[0]
bowled = df.loc[df['Dismissal'] == 'Bowled'].count()[0]
lbw = df.loc[df['Dismissal'] == 'LBW'].count()[0]
runout = df.loc[df['Dismissal'] == 'Run out'].count()[0]
stumped = df.loc[df['Dismissal'] == 'Stumped'].count()[0]
hitwicket = df.loc[df['Dismissal'] == 'Hit wicket'].count()[0]


plt.figure(figsize=(10,8))

labels = ['Caught', 'NO', 'Bowled', 'LBW', 'runout', 'stumped', 'hit wicket']
colors = ['#abcdef', '#aabbcc']

plt.pie([caught, no, bowled, lbw, runout, stumped, hitwicket], labels = labels, autopct='%.2f %%')

plt.title('Dismissals')

plt.show()
````
